### Functions

### What is a Function?

- A function is a reusable block of code that performs a specific task.
- Functions help avoid writing the same code multiple times.
- Once a function is created, it can be called whenever required.
- Functions improve code readability, organization, and reusability.
- The `def` keyword is used to define a function.

### Syntax

```python
def function_name():
    # code

````python
def greet():
    print("Hello, Welcome to Python")

greet()
````

### Function Execution

- Defining a function does not execute the code inside the function.
- When we create a function using the `def` keyword, Python only stores the function definition.
- The function runs only when it is called.
- The complete block of code inside the function executes during the function call.
- A function can be called multiple times, and each call executes the function again.



### Function Parameters and Arguments

- Parameters are variables defined when creating a function.
- Arguments are the actual values passed to the function when calling it.
- Parameters allow functions to work with different inputs.


### Print vs Return in Functions

### print()

- `print()` is used to display output on the screen.
- It only shows the result and does not send the value back.
- The printed value cannot be stored or reused in another operation.
- `print()` is mainly used for displaying information.

### return()

- `return` sends a value back from a function.
- The returned value can be stored in a variable and reused.
- A function can return different types of values such as numbers, strings, lists, or objects.
- `return` makes functions more flexible and reusable.
- Once a function reaches a return statement, the function execution stops.

 ### Prime Number Check Using Function

- A prime number is a number greater than 1 that has only two factors:
  - 1
  - The number itself
- Functions can be used to create reusable logic for checking whether a number is prime or not.
- A function can take a number as input and return True or False based on the condition.

### Iterators

### What is an Iterator?

- An iterator is an object that allows accessing elements one at a time.
- Iterators do not store all values in memory at once.
- They provide values sequentially when requested.
- Iterators are useful for processing large collections of data.

### iter()

- `iter()` is used to convert an iterable object into an iterator.
- Objects like lists, tuples, and strings can be converted into iterators.
- It returns an iterator object.


### next()

- `next()` is used to retrieve the next value from an iterator.
- Each call to `next()` moves the iterator to the next element.
- When there are no more elements, it raises a `StopIteration` error.

### Generators

### What is a Generator?

- A generator is a special type of iterator.
- It generates values one at a time instead of storing all values in memory.
- Generators are memory efficient and useful when working with large data.
- Generators are created using functions with the `yield` keyword.


### yield Keyword

- `yield` is used instead of `return` in generator functions.
- `yield` returns a value and pauses the function execution.
- The function state is saved after yielding a value.
- When called again, execution continues from where it stopped.

### Advantages of Generators

- Saves memory because values are generated only when needed.
- Useful for handling large datasets.
- Improves performance by avoiding unnecessary storage.
- Supports lazy evaluation, meaning values are created only when required.

### Fibonacci Series Using Generator

- The Fibonacci sequence is a series where each number is the sum of the previous two numbers.
- Generators are useful for creating Fibonacci sequences because values can be generated one by one.
- Instead of storing the complete sequence, generators produce each value when requested.


### Iterator vs Generator

| Iterator | Generator |
|----------|-----------|
| Created using `iter()` | Created using functions with `yield` |
| Requires creating an iterator object | Automatically creates an iterator |
| May require more code | Requires less code |
| Values are accessed using `next()` | Values are accessed using `next()` |
| Less memory efficient compared to generators | More memory efficient |

### map() Function

### What is map()?

- `map()` is a built-in Python function used to apply a function to every element in an iterable.
- It returns a map object containing the transformed values.
- It is useful when the same operation needs to be performed on multiple elements.

---

### Syntax 
map(function, iterable)

### Arguments of map()

`map()` takes two main arguments:

1. Function
   - The operation that needs to be performed.

2. Iterable
   - The collection of elements on which the function is applied.

Examples of iterables:
- Lists
- Tuples
- Strings



### reduce() Function

### What is reduce()?

- `reduce()` is used to combine multiple values into a single output value.
- It repeatedly applies a function to elements of a sequence.
- It reduces a collection into one final result.


### functools Module

- `reduce()` is not a built-in function in Python.
- It is imported from the `functools` module.


### reduce() with Lambda

- Lambda functions are commonly used with reduce().
- Lambda provides a short way to write simple functions.
- Reduce applies the lambda function repeatedly until one final value remains.


### filter() Function

### What is filter()?

- `filter()` is used to filter elements from an iterable based on a condition.
- It returns only the elements where the condition evaluates to True.
- It is commonly used for selecting specific data from a collection.


### filter() with Lambda

- `filter()` is commonly used with lambda functions.
- Lambda provides a simple way to write filtering conditions.
- Only values that satisfy the condition are returned.
