```python
# =====================================================
# PYTHON NOTES - JULY 21
# =====================================================

# =====================================================
# STRINGS
# =====================================================

# A string is an ordered sequence of characters enclosed
# within single quotes (' ') or double quotes (" ").

s = "Hello World"

# Built-in Function
# len() - Returns the total number of characters in a string.

len(s)
```

```python
# -----------------------------------------------------
# String Indexing and Slicing
# -----------------------------------------------------

# Indexing
# Indexing is used to access individual characters
# from a string using their position.

s[0]      # First character
s[-1]     # Last character

# Slicing
# Slicing is used to extract a portion of a string.

# Syntax:
# s[start:end:step]

s[0:5]    # Hello
s[::2]    # Every second character
s[::-1]   # Reverse string

# Reverse String
# Definition:
# Reversing a string means displaying characters
# in the opposite order.

s[::-1]
```

```python
# -----------------------------------------------------
# String Properties
# -----------------------------------------------------

# Concatenation - Combining two or more strings into a single string.
"Hello" + " World"

# Repetition - Repeats a string multiple times.
"Hi" * 3

# Immutability
# Strings cannot be modified after creation

s = "Hello"

# Item Reassignment
# Definition:
# Changing a value at a specific index.
# Not allowed in strings because strings are immutable.

# s[0] = 'h'  # Error
```

```python
# -----------------------------------------------------
# String Methods
# -----------------------------------------------------

s = "Hello World"

s.upper()      # HELLO WORLD, Converts all characters into uppercase.
s.lower()      # hello world, Converts all characters into lowercase.
s.split()      # ['Hello', 'World'], Converts all characters into lowercase.

# -----------------------------------------------------
# Print Formatting
# -----------------------------------------------------

# A modern way to insert variables directly into strings.

name = "John"

# f-string
print(f"My name is {name}")

# . format()
# Definition:
# Inserts values into placeholders within a string.

print("My name is {}".format(name))
```

```python
# -----------------------------------------------------
# Location and Counting Methods
# -----------------------------------------------------

s = "Hello World"

s.find("o")     # Returns the index position of the first occurrence of a specified value.
s.count("o")    # Counts how many times a value appears.
s.expandtabs()  # Replaces tab spaces with regular spaces.
```

```python
# -----------------------------------------------------
# String Validation Methods
# -----------------------------------------------------

s.isalnum()     # Returns True if string contains only letters and numbers.
s.isalpha()     # Returns True if string contains only alphabets.
s.islower()     # Returns True if all characters are lowercase.
s.isspace()     # Returns True if string contains only spaces.
s.istitle()     # Returns True if every word starts with a capital letter.
s.isupper()     # Returns True if all characters are uppercase.
s.endswith("!") # Ends with specified character

# Example

s = "Hello World!"

if s[-1] == '!':
    print("String ends with !")
else:
    print("String does not end with !")
```

```python
# -----------------------------------------------------
# Split and Partition
# -----------------------------------------------------

s = "Hello World"

s.split()          # Split into list
s.partition(" ")   # Split into 3 parts

# Splits a string into three parts:
# Before separator, separator, after separator.
```

```python
# =====================================================
# LISTS
# =====================================================

# List Basics
# List is an ordered collection that can hold multiple data types.
# Lists are mutable.

L = [1, "Python", 3.5, True]

# Length - Returns the number of elements in a list.
len(L)

# -----------------------------------------------------
# List Indexing
# -----------------------------------------------------

# Definition:
# Accessing individual elements using their position.

L[0]

# -----------------------------------------------------
# List Slicing
# -----------------------------------------------------

# Definition:
# Extracting a portion of a list.

L[1:3]
```

```python
# -----------------------------------------------------
# List Operations
# -----------------------------------------------------

# Concatenation 
# Combining two or more lists.
[1, 2] + [3, 4]

# Multiplication
# Repeating list elements multiple times.
[1, 2] * 3

# -----------------------------------------------------
# Mutability
# -----------------------------------------------------

# Lists can be modified after creation.
L = [1, 2, 2]

L[1] = 4

print(L)

# Output
# [1, 4, 2]
```

```python
# -----------------------------------------------------
# List Methods
# -----------------------------------------------------

L = [1, 2, 3]

L.append(4)      # Adds an element at the end of the list.

L.pop()          # Removes and returns the last element from a list. LIFO = Last In First Out

L.reverse()      # Reverse list 

L.sort()         # Sort ascending

# -----------------------------------------------------
# Advanced List Methods
# -----------------------------------------------------

L.count(2)       # Count occurrences

L.extend([5, 6]) # Add multiple values to a list

L.index(3)       # returns the position of a specified value

L.insert(1, 10)  # Insert a value at specified position

L.remove(2)      # Removes the first occurrence of specific value
```

```python
# -----------------------------------------------------
# Nested Lists
# -----------------------------------------------------

matrix = [
    [1, 2],
    [3, 4],
    [5, 6]
]

# List inside another list
```

```python
# -----------------------------------------------------
# List Comprehensions
# -----------------------------------------------------

# A compact way to create lists using a single line of code.
numbers = [x for x in range(10)]

squares = [x**2 for x in range(10)]

# Benefits:
# - Shorter code
# - Better readability
```

```python
# =====================================================
# TUPLES
# =====================================================

# Tuple is an ordered collection similar to list but immutable.

T = (1, 2, 3)

# Tuple to List- conerts tuple into list for modification

list(T)

# List to Tuple - converts list into tuple

L = [1, 2, 3]

tuple(L)

# When to use Tuple:
# - Permanent storage
# - when data should not be modified
```

```python
# =====================================================
# SETS
# =====================================================

# Set is an unordered collection of unique elements.
# Removes duplicates automatically.

S = {1, 2, 2, 3, 3}

print(S)

# Output
# {1, 2, 3}

# -----------------------------------------------------
# Set Methods
# -----------------------------------------------------

S.remove(1)      # Remove a specified element

S.discard(10)    # Removes an element without showing an error 

S.pop()          # Remove random element

S.clear()        # Remove all elements from the set
```

```python
# =====================================================
# DICTIONARIES
# =====================================================

# Dictionary stores data in key-value pairs. it is a mutable collection.

student = {
    "name": "John",
    "age": 22
}

# Access value

student["name"]

# Key
# Definition:
# Unique identifier used to access values.

# Value
# Definition:
# Data associated with a key.
```

```python
# -----------------------------------------------------
# Numerical Operations
# -----------------------------------------------------

marks = {
    "math": 90,
    "science": 85
}

sum(marks.values())

# -----------------------------------------------------
# Nested Dictionary
# -----------------------------------------------------

# A dictionary inside another dictionary
employee = {
    "emp1": {
        "name": "John",
        "age": 25
    }
}
```

```python
# -----------------------------------------------------
# Dictionary Methods
# -----------------------------------------------------

student.keys()      # Returns all keys in the dictionary

student.values()    # Returns all values in the dictionary

student.items()     # Returns key-value pairs as tuples

# -----------------------------------------------------
# Dictionary Comprehension
# -----------------------------------------------------

# A compact way to create dictionaries in a single line.
square_dict = { x: x**2 for x in range(5) }
```

```python
# =====================================================
# COPYING DICTIONARIES
# =====================================================

# Shallow Copy
# Copies only outer object, Creates a new object but shares nested objects
# with the original object.

d1 = {"a": 1, "b": 2}

d2 = d1.copy()

# Deep Copy
# Creates completely independent copy including all nested objects

import copy

d3 = copy.deepcopy(d1)
```

```python
# =====================================================
# QUICK COMPARISON
# =====================================================

# String
# Ordered      : Yes
# Mutable      : No
# Duplicates   : Yes
# Indexing     : Yes
# Syntax       : " "

# List
# Ordered      : Yes
# Mutable      : Yes
# Duplicates   : Yes
# Indexing     : Yes
# Syntax       : [ ]

# Tuple
# Ordered      : Yes
# Mutable      : No
# Duplicates   : Yes
# Indexing     : Yes
# Syntax       : ( )

# Set
# Ordered      : No
# Mutable      : Yes
# Duplicates   : No
# Indexing     : No
# Syntax       : { }

# Dictionary
# Ordered      : Yes
# Mutable      : Yes
# Key-Value Pairs
# Syntax       : {key:value}
```
