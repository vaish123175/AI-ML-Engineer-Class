```python
# ================================
# PYTHON BASICS NOTES
# ================================

# --------------------------------
# 1. WHAT IS PYTHON?
# --------------------------------
# Python is a high-level, interpreted, general-purpose programming language.
# High-level → Easy to read and write
# Interpreted → Executes line by line
# General-purpose → Used for web, data science, AI, etc.

print("Hello Python")
```

```python
# --------------------------------
# 2. PROCEDURAL EXECUTION
# --------------------------------
# Python executes code line by line from top to bottom.

a = 10
b = 20
sum_val = a + b
print(sum_val)   
```

```python
# --------------------------------
# 3. LIBRARIES
# --------------------------------
# A library is a collection of pre-written code that we can reuse.

import math
print(math.sqrt(25)) 
```

```python
# --------------------------------
# 4. VARIABLES
# --------------------------------
# A variable is used to store data.
# Variables are created using assignment (=)

name = "Vaishnavi"
age = 25
print(name)
```

```python
# --------------------------------
# 5. DYNAMIC TYPING
# --------------------------------
# Python automatically detects data type (no need to declare)

x = 10
print(type(x))   # int

x = "Hello"
print(type(x))   # str
```

```python
# --------------------------------
# 6. DATA TYPES
# --------------------------------
# Data types define the type of value stored.

# Integer (int) → whole numbers
num = 10

# Float → decimal numbers
price = 99.5

# Boolean → True or False
is_valid = True

# String → sequence of characters
text = "Python"

# Complex → real + imaginary
comp = 3 + 4j
```

```python
# --------------------------------
# 7. TYPE CASTING
# --------------------------------
# Converting one data type to another

# Implicit Casting → automatic conversion
a = 10
b = 5.5
c = a + b   # int → float automatically
print(c)

# Explicit Casting → manual conversion
x = "25"
x = int(x)
print(x + 5) 
```

```python
# --------------------------------
# 8. VARIABLE NAMING RULES
# --------------------------------
# 1. Cannot start with number
# 2. Cannot start with special character
# 3. Cannot use keywords (like class, def)
# 4. Should not use built-in names (like list, print)

student_name = "John"   # correct
```

```python
# --------------------------------
# 9. NAMING CONVENTION
# --------------------------------
# Snake case → words separated by underscore

total_salary = 50000
```

```python
# --------------------------------
# 10. OPERATORS
# --------------------------------

# Arithmetic Operators
a = 10
b = 3
print(a + b)   # Addition
print(a - b)   # Subtraction
print(a * b)   # Multiplication
print(a / b)   # Division
print(a % b)   # Modulus
print(a ** b)  # Power
print(a // b)  # Floor division


# Comparison Operators → return True/False
print(a > b)
print(a == b)
print(a != b)

# Logical Operators
# AND → both conditions must be True
print(a > 5 and b < 5)

# OR → at least one condition True
print(a > 5 or b > 10)

# NOT → reverse result
print(not True)


# Assignment Operators
x = 10
x += 5   # x = x + 5
x -= 3   # x = x - 3
print(x)

# Negative assignment
x = -5
```

```python
# --------------------------------
# 12. PRINT FUNCTION
# --------------------------------
# print() displays output

print("Welcome to Python")
```

```python
# --------------------------------
# 13. F-STRING
# --------------------------------
# Used to insert variables into strings

name = "John"
age = 25
print(f"My name is {name} and age is {age}")
```

```python
# --------------------------------
# 14. DOT FORMAT METHOD
# --------------------------------
print("My name is {} and age is {}".format(name, age))
```

```python
# --------------------------------
# 15. BOOLEAN EXPRESSIONS
# --------------------------------
# Expressions that return True or False

x = 10
print(x > 5)   # True

```

```python
# --------------------------------
# 16. BRANCHING (IF, ELIF, ELSE)
# --------------------------------

# IF
age = 20
if age >= 18:
    print("Adult")


# IF-ELSE
age = 15
if age >= 18:
    print("Adult")
else:
    print("Minor")

# IF-ELIF-ELSE
marks = 85
if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
else:
    print("Grade C")
```

```python
# --------------------------------
# 17. NESTED IF
# --------------------------------
# IF inside another IF

username = "admin"
password = "1234"

if username == "admin":
    if password == "1234":
        print("Login successful")
    else:
        print("Wrong password")
else:
    print("Wrong username")
```
