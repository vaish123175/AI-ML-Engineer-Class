```python
# =========================================
# CONDITIONAL STATEMENTS (Decision Making)
# =========================================

# Definition:
# Used to make decisions based on conditions

marks = 85

if marks >= 90:
    print("Grade A")   # runs if condition is true
elif marks >= 75:
    print("Grade B")   # runs if above condition is false and this is true
else:
    print("Grade C")   # runs if all above conditions are false
```

```python
# =========================================
# 📦 DATA COLLECTIONS (5 Types)
# =========================================

# 1. STRING (sequence of characters)
name = "Vaishnavi"
print(name[0])   # Access first character → V


# 2. LIST (ordered, mutable → can change)
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")   # adding element
print(fruits)


# 3. TUPLE (ordered, immutable → cannot change)
numbers = (1, 2, 3)
# numbers[0] = 10 ❌ Not allowed


# 4. SET (unordered, no duplicates)
nums = {1, 2, 2, 3}
print(nums)   # duplicates removed automatically


# 5. DICTIONARY (key-value pairs)
student = {"name": "Vaishnavi", "age": 22}
print(student["name"])   # access using key

```

```python
# =========================================
# 🔁 LOOPS (Repeat Actions)
# =========================================

# Definition:
# Used to perform an action multiple times

# Two types:
# 1. for loop
# 2. while loop
```

```python
# =========================================
# 🔹 FOR LOOP (known number of iterations)
# =========================================

for i in range(5):   # range starts from 0
    print(i)
```

```python
# =========================================
# 🔹 range() FUNCTION
# =========================================

# Syntax: range(start, stop, step)

# Example 1: numbers from 1 to 5
for i in range(1, 6):
    print(i)

# Example 2: step size = 2
for i in range(0, 10, 2):
    print(i)
```

```python
# =========================================
# 🔹 len() FUNCTION
# =========================================

# Definition:
# Returns number of elements

text = "python"
print(len(text))   # output: 6
```

```python
# =========================================
# 🔹 REVERSE STRING USING FOR LOOP
# =========================================

text = "python"
reverse = ""

for char in text:
    reverse = char + reverse   # add each char to front

print(reverse)   # output: nohtyp
```

```python
# =========================================
# 🔹 WHILE LOOP (condition-based loop)
# =========================================

# Definition:
# Runs until condition becomes False

i = 1

while i <= 5:
    print(i)
    i += 1   # increment to avoid infinite loop
```

```python
# =========================================
# 🔹 COUNTDOWN USING WHILE LOOP
# =========================================

count = 5

while count > 0:
    print(count)
    count -= 1   # decrease value

print("Time's up!")
```

```python
# =========================================
# 🔹 BREAK STATEMENT
# =========================================

# Definition:
# Stops the loop completely

for i in range(10):
    if i == 5:
        break   # loop stops here
    print(i)
```

```python
# =========================================
# 🔹 CONTINUE STATEMENT
# =========================================

# Definition:
# Skips current iteration and moves to next

for i in range(5):
    if i == 2:
        continue   # skip when i = 2
    print(i)
```
